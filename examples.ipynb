```python
import os
import pandas as pd
```

```python
def load_image(file_dir):
    file_path = os.listdir(file_dir)

    files = []
    labels = []

    for file in file_path:
        file_list = os.listdir(os.path.join(file_dir, file))
        files.extend(file_list)
        labels.extend([file] * len(file_list))
    
    df = pd.DataFrame({
    "file": files,
    "class": labels
    })

    df["file"] = file_dir + "/" + df["class"] + "/" + df["file"]

    return df
```

```python
train_df = load_image(file_dir = "Data/train")
test_df = load_image(file_dir = "Data/test")
```
